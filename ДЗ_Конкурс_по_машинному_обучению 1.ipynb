```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
!pip install catboost
from catboost import CatBoostClassifier
```

Разделение данных на тренировочную и валидационную выборки

```python
# YOUR CODE
X = data.iloc[:, :-1]
y = data['Churn']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.75, random_state=50, stratify=y) # поскольку имеется дисбаланс целевой переменной, то зададим параметр stratify
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

Обработка данных: One-hot кодирование

```python
# Создаем ColumnTransformer для масштабирования числовых и one-hot-encoding кодирования категориальных признаков
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False), cat_cols)
    ],
    remainder='passthrough')  # Оставить остальные столбцы без изменений
```

LogisticRegression

```python
# Создаем Pipeline, объединяющий предобработку и логистическую регрессию
pipeline = make_pipeline(
    preprocessor,
    LogisticRegression(random_state=50, max_iter=1000)
)

# Определяем гиперпараметры
parametrs = {'logisticregression__C': [1000, 100, 10, 1, 0.1, 0.01]}

# Создаем объект GridSearchCV для поиска наилучших параметров
grid_search_LR = GridSearchCV(pipeline, parametrs, cv=5, scoring='roc_auc')

# Обучаем модель
grid_search_LR.fit(X_train, y_train)

# Выведим наилучшие параметры и оценку ROC-AUC
print(grid_search_LR.best_params_)
print(grid_search_LR.best_score_)
```

```python
# Получаем вероятности и вычисляем AUC-ROC для валидационной выборки
probabilities_valid = grid_search_LR.predict_proba(X_valid)[:, 1]
LR_roc_auc = roc_auc_score(y_valid, probabilities_valid)

print("Логистическая регрессия:")
print("Значение ROC-AUC:", LR_roc_auc)
```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.



```python
# YOUR CODE
cbst_simple = CatBoostClassifier(random_state=50)

cbst_simple.fit(X_train, y_train, cat_features=cat_cols, early_stopping_rounds=30, verbose=False)

boost_preds_train = cbst_simple.predict_proba(X_valid)[:,1]
CBST_simple_roc_auc = roc_auc_score(y_valid, boost_preds_train)

print("CatBoost со стандартными настройками:")
print("Значение ROC-AUC:", CBST_simple_roc_auc)
```

2) Протестируйте разные значения параметров количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию.

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
cbst = CatBoostClassifier(random_state=50)

parametrs = {'learning_rate': [.2, .125, .1, .075],
        'iterations': [90, 100, 110, ]}

grid_search_catboost = GridSearchCV(cbst, parametrs, cv=5, scoring='roc_auc')

grid_search_catboost.fit(X_train, y_train, cat_features=cat_cols, verbose=False)

print(grid_search_catboost.best_params_)
print(grid_search_catboost.best_score_)
```

```python
boost_preds_train_simple = grid_search_catboost.predict_proba(X_valid)[:,1]
CBST_roc_auc = roc_auc_score(y_valid, boost_preds_train_simple)

print("CatBoost с подбором гиперпараметров:")
print("Значение ROC-AUC:", CBST_roc_auc)
```
